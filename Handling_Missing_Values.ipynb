```python
import numpy as np
import pandas as pd
import matplotlib as plt


#Dictionary of lists

dict = {'Mark' : [100,90,np.nan,95],
        'Rank': [1,2,4,np.nan],
        'Avg Marks': [np.nan,40,80,99]}

df = pd.DataFrame(dict)
df
```

```python
df.isnull()

#checks for missing value
```

```python
df.fillna(0)

#fillna is a function to fill the missing values with the value
```

```python
df.fillna(method='bfill')

#bfill is a function to fill the missing values with the next value
```

```python
df.fillna(method = 'pad')
#pad is a function to fill the missing values with the previous value
```

```python
df.fillna(method='ffill')
#ffill is a function to fill the missing values with the first value/
```
